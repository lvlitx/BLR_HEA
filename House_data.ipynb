```python
# check the values in BHK if they are float 
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
data[~data.total_sqft.apply(is_float)].head()
```

```python
#creating a function to convert sqft to number 
def sqft_to_num(x):
    
    tokens = x.split("-")
    
    if len(tokens) == 2 :
        return np.average([float(tokens[0]),float(tokens[1])])

    try :
        return float(x)
    
    except:
        return None
        
```

```python
# check the function 
sqft_to_num('200-300')
```

```python
# applying the functin to every thing so that it converts into the number 
data['total_sqft'] = data['total_sqft'].apply(sqft_to_num)
```

```python
# checking how much data is available for each location 
data.location = data.location.apply(lambda x : x.strip())

stats_of_data = data.groupby('location')['location'].agg('count') 

stats_of_data.sort_values(ascending = False)
```

```python
#check if we have location that has 10 or less values 
len(stats_of_data[stats_of_data <= 10])
```

```python
# modifiying the data if the with "other" if they are less than 10
data.location = data.location.apply(lambda x :'other' if x in stats_of_data else x  )
# this lambda function will replace the values that are lessthan 10 with other 
```
